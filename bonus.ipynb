```python
bonus = open('C:/Users/naiem/OneDrive/Desktop/44-608/608-mod4/bonus.txt', 'r')

#  use the forward-slash ('/') even if youâ€™re working on Windows
```

```python
text = bonus.read()
```

```python
word_counts = {}
```

```python
# count occurrences of each unique word
for word in text.split():
    if word in word_counts: 
        word_counts[word] += 1
    else:
        word_counts[word] = 1
```

```python
print(f'{"WORD":<12}COUNT')
for word, count in sorted(word_counts.items()):
    print(f'{word:<12}{count}')
```
