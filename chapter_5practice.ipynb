# 5.14- Filter, Map and Reduce

```python
numbers = [10, 3, 7, 1, 9, 4, 2, 8, 5, 6]
```

```python
def is_odd(x):
    """Returns True value only if x is odd."""
    return x % 2 != 0
    
```

```python
list(filter(is_odd, numbers))
```

<b>Using a lambda Rather than a Function</b>

```python
list(filter(lambda x: x % 2 != 0, numbers))
```

<b>Mapping a Sequence's Values to New Values</b>

```python
numbers
```

```python
list(map(lambda x: x ** 2, numbers))
```

```python
[item **2 for item in numbers]
```

<b> Combining filter and map</b>

```python
list(map(lambda x: x ** 2,
         filter(lambda x: x % 2 != 0, numbers)))
```

```python
[x ** 2 for x in numbers if x % 2 != 0]
```

# 5.16 - Two-Dimensional Lists

<b>Creating a Two-Dimensional List</b>

```python
a = [[77, 68, 86, 73], [96, 87, 89, 81], [70, 90, 86, 81]]
```

<b> Identifying the Elements in a Two-Dimensional List </b>

```python
for row in a:
    for item in row:
        print(item, end=' ')
    print()
```

<b>How the Nested Loops Execute</b>

```python
for i, row in enumerate(a):
    for j, item in enumerate(row):
        print(f'a[{i}][{j}] ={item} ', end=' ')
    print()
```
